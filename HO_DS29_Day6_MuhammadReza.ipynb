## Tipe Data Lanjutan

```python
# Dictionary
adalah tipe data yang dapat menyimpan sebuah value dengan menandainya dengan sebuah key.
``
{key:value}
``
- key pada dictionary bersifat unique
``
{"siswa":[1,2,3,4,5],
 "siswa_1":[5,6,7,8,9]}
```

```python
profile_student = {
    "nama": "reza"
}
profile_student
```

```python
profile_student = { 
    "nama":["reza", "cipi", "silvi"]
}

profile_student
```

```python
# cara membuat key yang baru
profile_student["kelas"] = ["A", "B", "C"]
profile_student
```

```python
# membuat profile student yang lengkap
profile_student = {
    "nama":["Reza", "Silvia", "Radya", "Heyzel", "Putri"],
    "nim":["1234", "1342", "2341", "3421", "4231"],
    "gender":["Pria","Wanita","Pria","Pria","Wanita"],
    "jurusan":["Teknik Informatika", "Kesehatan Lingkungan", "Akuntansi", "Manajemen", "Hukum"],
}

print(profile_student)
```

# Cara mengakses Value berdasarkan Key tertentu

`get(key)` untuk mengakses value tertentu dari sebuah Key<br>
`dictionary[key]` untuk mengakses value tertentu dari Key<br>

```python
profile_student.get("jurusan")
```

```python
profile_student.get("nama")
```

# Mengakses banyak Key dengan Looping FOR

`.keys()` adalah sebuah function yang digunakan untuk mengembalikan key yang ada pada dictionary yang telah didefinisikan

```python
profile_student.keys()
```

```python
for key in profile_student.keys():
    print(f"Daftar {key} : {', '.join(profile_student.get(key))}")
```

`values()` adalah sebuah function yang hanya akan mengembalikan values yang ada pada dictionary yang telah didefinisikan

```python
profile_student.values()
```

```python
for value in profile_student.values():
    print(", ".join(value))
```

`items` sebuah function yang mengembalikan (key, value) pada dictionary yang telah didefinisikan

```python
profile_student.items()
```

```python
for key, value in profile_student.items():
    print(f"Daftar {key} : {', '.join(profile_student.items(key))}")
```
